```python
import cv2
import easygui
from matplotlib import pyplot as plt
from matplotlib import image as image

I1 = cv2.imread('orange.png')
I2 = cv2.imread('water.jpg')
I1 =cv2.cv2.cvtColor(I1,cv2.COLOR_BGR2RGB)
I2 =cv2.cv2.cvtColor(I2,cv2.COLOR_BGR2RGB)


h,w,d =I1.shape
print(h,w,d)
h = int(h*2)
w = int(w*2)

I1=cv2.resize(I1,dsize=(w,h))
h,w,d =I1.shape
print(h,w,d)

plt.axis('off')
plt.imshow(I1)
plt.show()

h,w,d =I2.shape
print(h,w,d)
h = int(h*2)
w = int(w*2)

I2=cv2.resize(I2,dsize=(w,h))
h,w,d =I2.shape
print(h,w,d)

plt.axis('off')
plt.imshow(I2)
plt.show()

A =cv2.add(I1,I2)

plt.axis('off')
plt.imshow(A)
plt.show()




```
