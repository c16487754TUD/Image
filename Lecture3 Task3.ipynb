```python
import cv2
import easygui
from matplotlib import pyplot as plt
from matplotlib import image as image
import numpy as np

Orange = cv2.imread('orange.png')
Orange =cv2.cv2.cvtColor(Orange,cv2.COLOR_BGR2RGB)

Water = cv2.imread('water.jpg')
Water =cv2.cv2.cvtColor(Water,cv2.COLOR_BGR2RGB)


print("Original Images:")
plt.axis('off')
plt.imshow(Orange)
plt.show()

plt.axis('off')
plt.imshow(Water)
plt.show()

RangeLower=(150,100,0)
RangeUpper=(255,250,250)
Mask =cv2.inRange(Orange,RangeLower,RangeUpper)

ROI = cv2.bitwise_and(Orange,Orange,mask=Mask)
ReverseMask = cv2.bitwise_not(Mask)

Mask2 = cv2.bitwise_and(Water,Water,mask=ReverseMask)

Final = cv2.bitwise_or(ROI,Mask2)

print("Final Image:")
plt.axis('off')
plt.imshow(Final)
plt.show()


```
