```python
import cv2
import easygui
from matplotlib import pyplot as plt
from matplotlib import image as image

I =cv2.imread('image.jpg')
I =cv2.cv2.cvtColor(I,cv2.COLOR_BGR2RGB)
I2 =cv2.cv2.cvtColor(I,cv2.COLOR_BGR2YUV)
I3 =cv2.cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)

print("Original Image:")
plt.axis('off')
plt.imshow(I)
plt.show()


print("YUV Image(only y channel):")
Y = I2[:,:,0]
plt.axis('off')
plt.imshow(Y)
plt.show()

print("grayscale image:")
plt.axis('off')
plt.imshow(I3, cmap = "gray")
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Side by side:')
ax1.axis('off')
ax1.imshow(Y)
ax2.axis('off')
ax2.imshow(I3, cmap = "gray")
fig.tight_layout(pad=1.0)


```
