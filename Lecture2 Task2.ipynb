```python
import cv2
import easygui
from matplotlib import pyplot as plt
from matplotlib import image as image

I = cv2.imread('googly.jpg')
h,w,d =I.shape
            
x1= int(w/2)
x2=int(w)
y1=int(h/2)
y2=int(h)

C =I[y1:y2, x1:x2]
h,w,d =C.shape
cx = w/2
cy = h/2


M =cv2.getRotationMatrix2D(center=(cx,cy),angle=45, scale=.5)
R =cv2.warpAffine(C, M = M,dsize=(w,h))


plt.axis('off')
plt.imshow(R)
plt.show()
```
