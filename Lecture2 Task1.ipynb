```python
import cv2
import pylab as plt
import matplotlib.image as mpimg
import numpy as np


G = plt.imread('Wartime.jpg')
G =cv2.cvtColor(G,cv2.COLOR_BGR2GRAY)

Values = G.ravel()
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Before:')
ax1.axis('off')
ax1.imshow(G, cmap = "gray")
ax2.hist(Values,bins=256,range=[0,256]); 
fig.tight_layout(pad=1.0)

H = cv2.equalizeHist(G)
Values = H.ravel()
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('After:')
ax1.axis('off')
ax1.imshow(H, cmap = "gray")
ax2.hist(Values,bins=256,range=[0,256]); 
fig.tight_layout(pad=1.0)
```
